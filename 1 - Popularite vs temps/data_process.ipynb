```python
import pandas as pd
import altair as alt
import plotly.express as px
import numpy as np
```

```python
df = pd.read_csv("../data/dpt2020.csv", sep=";")
df = df[df["preusuel"] != "_PRENOMS_RARES"]
df = df[df["annais"] != "XXXX"]
df = df.sort_values("annais")
df = df.drop("sexe", axis=1)

data = pd.DataFrame({
    "prenom": df["preusuel"],
    "year": df["annais"],
    "departement": df["dpt"],
    "nombre": df["nombre"]
})
data
```

```python
prenoms = data.groupby("prenom").sum()
prenoms = prenoms.sort_values("nombre", ascending=False)
prenoms = prenoms[0:200]
# prenoms
```

```python
prenom_per_year = data.groupby("year").sum().reset_index()
prenom_per_year
```

```python
years = data["year"].unique()
print(len(years), "années")
print(len(prenoms), "prenoms")

prenoms_col, years_col, nb, percent = [], [], [], []
for prenom in prenoms.index:
    df_ = data[data["prenom"] == prenom]
    for year in years:
        cc = df_[df_["year"] == year]["nombre"].sum()
        prenoms_col.append(prenom)
        years_col.append(int(year))
        nb.append(cc)
        percent.append(cc / prenom_per_year[prenom_per_year["year"] == year].values[0][1] * 100)

dwh = pd.DataFrame({
    "prenom": prenoms_col,
    "year": years_col,
    "nombre": nb,
    "parts": percent
})
dwh
```

```python
dwh.to_csv("data_part_1.csv", index=False)
```
